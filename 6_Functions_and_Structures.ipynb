## 1. Khái niệm

### Lợi ích của hàm

## 2. Hàm `main()`

- Một chương trình C phải có **duy nhất 1 hàm `main()`**
- `main()` là nơi chương trình bắt đầu chạy

```c
int main() {
    // các lệnh xử lý của hàm main
    return 0;
}
```

- Kiểu trả về của `main` thường là `int`
- Giá trị trả về:
  - `0` hoặc `EXIT_SUCCESS`: chương trình kết thúc thành công
  - `1` hoặc `EXIT_FAILURE`: chương trình kết thúc có lỗi

````python
%%bash
cat << 'EOF' > main_demo.c
#include <stdio.h>
#include <stdlib.h>

int main() {
    printf("Program starts here.\n");
    return 0;
}
EOF

gcc main_demo.c -o main_demo
./main_demo
````

## 4. Một số hàm toán học trong `<math.h>`

Các hàm toán học nằm trong thư viện `math.h`.

## 8. Biến `static`

- Biến `static` cục bộ **vẫn giữ được giá trị trước đó qua các lần gọi hàm**
- Cú pháp:

```c
static <type> <variable> [= <ConstExpr>];
```

Ví dụ:

```c
void TT() {
    static int a = 1;
    printf("%d\t", a++);
}
```

- Nếu không có `static`, biến cục bộ sẽ bị xóa sau mỗi lần gọi hàm

````python
%%bash
cat << 'EOF' > static_demo.c
#include <stdio.h>

void TT() {
    static int a = 1;
    printf("%d\t", a++);
}

int main() {
    TT();
    TT();
    TT();
    printf("\n");
    return 0;
}
EOF

gcc static_demo.c -o static_demo
./static_demo
````

## 9. Nguyên tắc thực thi khi gọi hàm

Khi gọi hàm, bộ thực thi sẽ:

2. Copy các thông số cho hàm được gọi
3. Chuyển điều khiển thực thi cho hàm được gọi
4. Khi gặp `return`:
   - giải phóng các biến cục bộ của hàm
   - trả điều khiển về lệnh sau lời gọi hàm

## 10. Truyền tham số bằng giá trị (`int a`, `float b`,...)
Biến gửi vào **không bị thay đổi giá trị** sau khi thực hiện hàm

Ví dụ `swap` sai do truyền bằng giá trị:

```c
void swap(int a, int b) {
    int t = a;
    a = b;
    b = t;
}
```

````python
%%bash
cat << 'EOF' > swap_value.c
#include <stdio.h>

void swap(int a, int b) {
    int t = a;
    a = b;
    b = t;
}

int main() {
    int x = 10, y = 100;
    printf("Before swap(x, y): x = %d, y = %d\n", x, y);
    swap(x, y);
    printf("After swap(x, y):  x = %d, y = %d\n", x, y);
    return 0;
}
EOF

gcc swap_value.c -o swap_value
./swap_value
````

### Nhận xét

- `x` và `y` **không đổi**
- Vì hàm chỉ nhận **bản sao giá trị**

## 11. Truyền tham số bằng địa chỉ (pointers - `int *a`, `char *b`,...)

- Biến gửi vào **có thể bị thay đổi**
- Tham số hình thức có dạng:

```c
(type *tên)
```

- Tham số thực có dạng:

```c
(&tên)
```

### Lưu ý

Tham số truyền vào phải là **biến**.

Không thể truyền:

```c
// invalid
swap(10, 100);
swap(x + 10, y + 2);
```

vì đó không phải là địa chỉ của biến.

## Truyền tham số bằng địa chỉ – ví dụ `swap`

```c
void swap(int *a, int *b) {
    int t = *a;
    *a = *b;
    *b = t;
}
```

````python
%%bash
cat << 'EOF' > swap_address.c
#include <stdio.h>

void swap(int *a, int *b) {
    int t = *a;
    *a = *b;
    *b = t;
}

int main() {
    int x = 10, y = 100;
    printf("Before swap(&x, &y): x = %d, y = %d\n", x, y);
    swap(&x, &y);
    printf("After swap(&x, &y):  x = %d, y = %d\n", x, y);
    return 0;
}
EOF

gcc swap_address.c -o swap_address
./swap_address
````

## 13. Hàm đệ quy

- Hàm đệ quy là hàm trong thân hàm có lệnh gọi lại chính nó

### Các loại đệ quy

- **Trực tiếp**: `foo()` gọi `foo()`
- **Gián tiếp**: `foo()` gọi `bar()`, `bar()` lại gọi `foo()`

## Ý tưởng của đệ quy

Có thể xem đệ quy như một vòng lặp ở cấp độ hàm.

Một số công thức quen thuộc:

- `n! = (n-1)! × n`
- `S(n) = S(n-1) + n`
- `Fibo(n) = Fibo(n-1) + Fibo(n-2)`

## Ví dụ đệ quy – tính giai thừa

```c
int giai_thua(int n) {
    if (n <= 1)
        return 1;
    else
        return n * giai_thua(n - 1);
}
```

### Điểm dừng (base case)

- Khi `n <= 1`, hàm trả về `1`
- Nếu không có điểm dừng, đệ quy sẽ vô hạn

````python
%%bash
cat << 'EOF' > factorial_recursive.c
#include <stdio.h>

int giai_thua(int n) {
    if (n <= 1)
        return 1;
    else
        return n * giai_thua(n - 1);
}

int main() {
    printf("5! = %d\n", giai_thua(5));
    return 0;
}
EOF

gcc factorial_recursive.c -o factorial_recursive
./factorial_recursive
````

## Ghi nhớ về đệ quy

Khi gọi đệ quy, giá trị đầu vào thường phải **giảm dần** hoặc tiến dần về điều kiện dừng.

Ví dụ:

- `ex(0) = 1`
- `x^0 = 1`
- `1! = 1`

## 14. Tổ chức chương trình C

### Thành phần thường gặp

- **Tiền xử lý**
  - `#include`
  - `#define`
- **Khai báo / định nghĩa**
  - hằng
  - biến
  - hàm / biến `extern`
- **Mô tả hàm**
- **Các hàm**
  - `main` chạy đầu tiên
  - các hàm được định nghĩa ngang cấp, không lồng nhau
  - được gọi thì mới chạy

## Tổ chức mã nguồn trong chương trình

Ví dụ:

```c
#include <stdio.h>
#include <stdlib.h>

int add(int a, int b);

int main() {
    printf("10 + 15 = %d", add(10, 15));
    return 0;
}

int add(int a, int b) {
    int c;
    c = a + b;
    return c;
}
```

### Ý nghĩa

- Tách phần **mô tả hàm** (prototype) đặt trước `main`
- Cài đặt phần thân hàm phía sau

## 15. Tổ chức mã nguồn chương trình lớn

### Tách thành nhiều tập tin

- **Header file**: `*.h`
  - chứa mô tả hàm, kiểu dữ liệu, hằng...
- **Implementation file**: `*.c`
  - chứa phần hiện thực
- **Main program file**
  - chứa `main()`
  - gọi các hàm

## Header file và include guard

Ví dụ `my_math.h`:

```c
#if !defined(MY_MATH_HEADER)
#define MY_MATH_HEADER

extern int add(int, int);

#endif
```

### Ý nghĩa

- Tránh lỗi định nghĩa lặp lại (`redefinition`)
- Nếu macro chưa được định nghĩa thì biên dịch phần nội dung bên trong

## Implementation file

Ví dụ `my_math.c`:

```c
int add(int a, int b) {
    int c;
    c = a + b;
    return c;
}
```

Có thể khai báo biến ở module khác bằng:

```c
extern <type> <variable>;
```

## Main program file

Ví dụ `program.c`:

```c
#include <stdio.h>
#include <stdlib.h>
#include "my_math.h"

int main() {
    printf("10 + 15 = %d", add(10, 15));
    return 0;
}
```

````python
%%bash
cat << 'EOF' > my_math.h
#if !defined(MY_MATH_HEADER)
#define MY_MATH_HEADER

int add(int a, int b);

#endif
EOF

cat << 'EOF' > my_math.c
int add(int a, int b) {
    return a + b;
}
EOF

cat << 'EOF' > program.c
#include <stdio.h>
#include "my_math.h"

int main() {
    printf("10 + 15 = %d\n", add(10, 15));
    return 0;
}
EOF

gcc program.c my_math.c -o program
./program
````
