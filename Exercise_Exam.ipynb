# Đáp án kèm giải thích ngắn

1. **D** — biểu thức vừa dùng `x++` vừa `++x` trên cùng biến trong một biểu thức là tình huống không nên kết luận chắc chắn; đây là bẫy về thứ tự đánh giá / side effects.
2. **A** — `s[1] = 'e'`, còn `*(s + 4) = 'o'`.
3. **C** — `p = a + 1` trỏ tới `a[1]`, nên `*(p + 2) = a[3] = 8`.
4. **C** — truyền địa chỉ nên hàm sửa trực tiếp biến gốc, `a` thành 17.
5. **A** — `x` truyền bằng giá trị nên chỉ đổi bản sao, `a` vẫn là 9.
6. **C** — vào `case 3` rồi rơi tiếp xuống `case 5` và `default` vì không có `break`.
7. **A** — in `0 1`, đến `i == 2` thì `break`.
8. **B** — `continue` bỏ qua các số chẵn, nên chỉ in `1 3 5`.
9. **C** — đây là giai thừa: `4 * 3 * 2 * 1 = 24`.
10. **C** — `g(8)=1+g(4)=1+1+g(2)=1+1+1+g(1)=1+1+1+1+g(0)=4`.
11. **A** — mảng ký tự kết thúc bằng `'\0'`, nên in trọn chuỗi `ABC`.
12. **C** — `pp` trỏ tới `p`, còn `**pp` là giá trị cuối cùng của `a`, tức 10.
13. **B** — `*(a + 1)` chính là `a[1] = 2`.
14. **B** — toán tử `?:` chọn giá trị lớn hơn, là 8.
15. **C** — với nhiều đối số trong `printf`, không nên dạy học sinh giả định chắc thứ tự đánh giá đối số; đây là câu bẫy nhận thức tốt.
16. **B** — hàm `swap` dùng pointer nên đổi thật giá trị của `x` và `y`.
17. **B** — `p += 3` làm `p` trỏ tới `a[3]`, nên `*p = 3`.
18. **A** — `ptr->x + ptr->y = 2 + 7 = 9`.
19. **B** — `%05d` in số nguyên độ rộng 5, thêm số 0 ở đầu.
20. **B** — `x / 2` là chia nguyên trước, `9 / 2 = 4`, rồi mới gán sang `float` thành `4.0`.

# Đáp án kèm giải thích ngắn

1. **B** — trong tin học cơ bản, `1 byte = 8 bit`.
2. **C** — `char` thường chiếm 1 byte và chủ yếu dùng để lưu ký tự.
3. **C** — `sizeof()` dùng để lấy số byte của kiểu dữ liệu hoặc biến.
4. **B** — `do...while` kiểm tra điều kiện sau nên luôn chạy ít nhất 1 lần; `while` kiểm tra trước nên có thể không chạy lần nào.
5. **A** — `switch` thường làm việc với giá trị rời rạc như `int`, `char`, `enum`; nếu thiếu `break` có thể bị fall-through.

7. **C** — `fclose()` dùng để đóng file sau khi làm việc xong.
8. **D** — `fscanf()` đọc dữ liệu có định dạng từ file, tương tự `scanf()` nhưng áp dụng cho file.
9. **A** — `fprintf()` ghi dữ liệu có định dạng ra file, tương tự `printf()` nhưng áp dụng cho file.
10. **B** — `NULL` nghĩa là con trỏ chưa trỏ đến vùng nhớ hợp lệ nào, nên không được dereference trực tiếp.
