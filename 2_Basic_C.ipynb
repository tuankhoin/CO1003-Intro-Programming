## Chương trình C đầu tiên

#### Chú thích (Comments)
Được chèn vào nhằm giải thích, ghi chú. Trình biên dịch sẽ bỏ qua các dòng này.

Có 2 dạng:
- chú thích 1 dòng `// comment`
- chú thích nhiều dòng `/* comment */`

#### `#include`
- Khai báo cho bộ tiền xử lý biết phải nạp nội dung của thư viện `<xxx.h>` vào trước khi biên dịch
- Thư viện nhập xuất chuẩn trong C là `<stdio.h>` (Standard Input/Output). Có 2 hàm từ đây:
- `printf()`: hàm xuất có định dạng
- `scanf()`: hàm nhập có định dạng

#### `main()`

Chương trình C có thể có nhiều hàm, tuy nhiên bắt buộc  phải có một và chỉ một hàm `main()`. Chương trình sẽ bắt đầu chạy từ hàm này.



````python
%%bash
cat << 'EOF' > comment.c
////////////////////////////////////////
#include <stdio.h>

int main() {
    // Chú thích một dòng
    printf("Hello\n");  // In ra Hello

    /*
       Chú thích
       nhiều dòng
    */
    printf("Welcome to C!\n");
    return 0;
}
////////////////////////////////////////
EOF
gcc comment.c -o comment
./comment
````

## Escape sequence và printf

Escape sequence bắt đầu bằng dấu \ theo sau là 1 ký tự có ý nghĩa riêng, để xác định 1 hành động đặc biệt.  Ví dụ: \n là xuống dòng và ra đầu dòng.

| Escape Sequence | Meaning                          |
|-----------------|----------------------------------|
| \\              | \ character                      |
| \'              | ' character                      |
| \"              | " character                      |
| \?              | ? character                      |
| \a              | Alert or bell                    |
| \b              | Backspace                        |
| \n              | Newline                          |
| \r              | Carriage return                  |
| \t              | Horizontal tab                   |
| \ooo            | Octal number (1 to 3 digits)     |
| \xhh...         | Hexadecimal number (1+ digits)   |


````python
%%bash
cat << 'EOF' > escape.c
////////////////////////////////////////
#include <stdio.h>

int main() {
    printf("Welcome ");
    printf("to C!\n");
    printf("Welcome\nto\nC!\n");
    return 0;
}
////////////////////////////////////////
EOF
gcc escape.c -o escape
./escape
````

## `printf` với nhiều kiểu dữ liệu

`printf("<chuỗi định dạng>", <danh sách tham số>);`

Chỉ thị định dạng trong chuỗi định dạng có dạng bắt đầu bằng dấu % theo sau là specifier, với specifier đặc tả cho kiểu dữ liệu

| Format Specifier | Meaning                                 | Example (Value → Output)      |
|------------------|-----------------------------------------|-------------------------------|
| %d or %i         | Signed decimal integer                  | 42 → 42                       |
| %f or %g         | Floating-point number                   | 3.14159 → 3.141590            |
| %e               | Floating-point in exponential notation  | 3.14159 → 3.141590e+00        |
| %x or %X         | Hexadecimal integer                     | 255 → ff (or FF)              |
| %o               | Octal integer                           | 255 → 377                     |
| %c               | Single character                        | 'A' → A                       |
| %s               | String of characters                    | "Hello" → Hello               |


- Ta có thể quy định độ rộng của dữ liệu in phía trước ký tự đặc tả kiểu. Ví dụ: `%8d`, `10.2f`, `%20s`
- Dấu chấm, dấu trừ, ... Đều được tính là chiếm 1 vị trí
- Dữ liệu in ra mặc định canh lề phải. Muốn canh trái ta sử dụng thêm dấu trừ. Ví dụ “%-8d”
- Nếu dữ liệu in ra lớn hơn kích thước mô tả thì sẽ tự động dãn ra để in đủ dữ liệu


````python
%%bash
cat << 'EOF' > printf_types.c
////////////////////////////////////////
#include <stdio.h>

int main() {

    int a=4, b=5;
	  printf("Ket qua: %d + %d = %d \n",a, b, a+b);

    // In nhiều biến 1 lần
    printf("1. %04d\t%5.4f\n", 123, 123.45);
    printf("2. %c\t%s\n", 'A', "Hello");

    // In nhiều kiểu dữ liệu khác nhau
    printf("3. %d\n", 123);
    printf("4. %5d\n", 123);
    printf("5. %05d\n", 123);

    printf("6. %f\n", 123.456);
    printf("7. %10.2f\n", 123.456);
    printf("8. %010.2f\n", 123.456);

    printf("9. %3c\n",'b');

    printf("10. %s\n","1234567890");
    printf("11. %s\n","DHBK");
    printf("12. %9s\n","DHBK");

    return 0;
}
////////////////////////////////////////
EOF
gcc printf_types.c -o printf_types
./printf_types
````

## Nhập dữ liệu với scanf
`scanf ("<chuỗi định dạng>" , <địa chỉ biến>)`

````python
%%bash
cat << 'EOF' > scanf_one.c
////////////////////////////////////////
#include <stdio.h>

int main() {
    int x;            // Khai báo biến và định dạng
    printf("Please enter x: \n");
    scanf("%d", &x); // &x là địa chỉ của biến x
    printf("x = %d\n", x);
    return 0;
}
////////////////////////////////////////
EOF
gcc scanf_one.c -o scanf_one
echo "123" | ./scanf_one
````

## Phép toán số nguyên và số thực

| Phép toán         | Toán tử | Biểu thức đại số | Biểu thức trong C |
|-------------------|---------|------------------|--------------------|
| Phép cộng         | +       | f + 7            | f + 7              |
| Phép trừ          | -       | p - c            | p - c              |
| Phép nhân         | *       | b . m            | b * m              |
| Phép chia         | /       | x / y            | x / y              |
| Phép lấy phần dư  | %       | r mod s          | r % s              |

- Phép chia số nguyên cho ra kết quả là phần thương của hai số.
  - 7 / 4 cho ra 1  
  (nếu viết 7.0 / 4 hoặc 7 / 4.0 hoặc 7.0 / 4.0 kết quả sẽ bằng 1.75)
  - 17 / 5 cho ra 3
- Phần thập phân bị **cắt đi**, chứ không phải làm tròn
- Phép chia lấy số dư (%) : hai toán hạng là số nguyên


````python
%%bash
cat << 'EOF' > division.c
////////////////////////////////////////
#include <stdio.h>

int main() {
    float x;
    x = 7 / 4;
    printf("7 / 4 = %f\n", x);

    x = 7.0 / 4;
    printf("7.0 / 4 = %f\n", x);
    return 0;
}
////////////////////////////////////////
EOF
gcc division.c -o division
./division
````

## `struct` - Dữ liệu cấu trúc

Là kiểu dữ liệu cho phép kết hợp các dữ liệu khác kiểu nhau. Sẽ học kỹ ở chương sau.

````python
%%bash
cat << 'EOF' > struct_demo.c
////////////////////////////////////////
#include <stdio.h>

struct Book {
    char title[50];
    int book_id;
};

int main() {
    struct Book b = {"C Programming", 101};
    printf("Title: %s\n", b.title);
    printf("ID: %d\n", b.book_id);
    return 0;
}
////////////////////////////////////////
EOF

gcc struct_demo.c -o struct_demo
./struct_demo
````

# Quy chuẩn đặt tên biến và hàm trong ngôn ngữ C

Cảm ơn ChatGPT

## 1. Nguyên tắc chung

Tên biến và tên hàm không chỉ để **máy hiểu**, mà chủ yếu để **con người đọc và hiểu**.
Một chương trình tốt là chương trình mà người khác (hoặc chính bạn sau 6 tháng) đọc lại **không cần đoán ý đồ**.

Quy tắc chung:

* Tên phải **có ý nghĩa**
* Tránh viết tắt mơ hồ
* Nhất quán trong toàn bộ chương trình
* Ưu tiên dễ đọc hơn ngắn gọn

---

## 2. Quy chuẩn đặt tên hàm (Function Naming)

### 2.1. Hàm nên bắt đầu bằng **động từ**

Hàm là **đơn vị xử lý**, vì vậy tên hàm nên thể hiện **hành động**.

✅ Nên:

```c
check_for_errors()
calculate_total_price()
read_input()
print_result()
```

❌ Không nên:

```c
error()
total()
input()
result()
```

---

### 2.2. Quy ước tiền tố (prefix) thường dùng cho hàm

Một số tiền tố giúp tăng ngữ nghĩa và khả năng đọc hiểu:

| Prefix   | Ý nghĩa                                | Ví dụ               |
| -------- | -------------------------------------- | ------------------- |
| `is_`    | Kiểm tra điều kiện (trả về true/false) | `is_prime_number()` |
| `get_`   | Lấy giá trị                            | `get_user_age()`    |
| `set_`   | Gán giá trị                            | `set_motor_speed()` |
| `init_`  | Khởi tạo                               | `init_system()`     |
| `print_` | Xuất dữ liệu                           | `print_menu()`      |

---

### 2.3. Kiểu chữ và cách nối từ

Hai cách phổ biến, **chọn 1 và dùng xuyên suốt**:

* `snake_case` (khuyến nghị trong C)

```c
calculate_average_score()
read_sensor_data()
```

* `camelCase` (ít dùng hơn trong C)

```c
calculateAverageScore()
readSensorData()
```

---

## 3. Quy chuẩn đặt tên biến (Variable Naming)

### 3.1. Biến nên bắt đầu bằng **danh từ**

Biến đại diện cho **dữ liệu**, không phải hành động.

✅ Nên:

```c
int student_count;
float average_score;
char user_name[50];
```

❌ Không nên:

```c
int calculate;
float check;
char print[50];
```

---

### 3.2. Kiểu chữ và nối từ

* Dùng **chữ thường**
* Nối các từ bằng dấu gạch dưới `_`

```c
int total_price;
int number_of_students;
float motor_speed;
```

---

### 3.3. Biến con trỏ (Pointer)

* Dấu `*` đặt **sát tên biến**
* Có thể dùng hậu tố `_ptr` để rõ nghĩa

```c
char *name;
int *count_ptr;
struct Student *student_ptr;
```

---

## 4. Biến toàn cục và hằng số

### 4.1. Biến toàn cục (Global variables)

* Dùng tiền tố `g_` để phân biệt với biến cục bộ

```c
int g_total_students;
float g_system_voltage;
```

---

### 4.2. Hằng toàn cục (Global constants)

* Viết **toàn bộ chữ hoa**
* Các từ nối bằng `_`
* Thường khai báo với `const`

```c
const int MAX_STUDENTS = 50;
const float PI = 3.1415926;
```

## 5. Ví dụ tổng hợp (ĐÚNG chuẩn)

```c
#include <stdio.h>

const int MAX_SCORE = 100;

    return score >= 50;
}

void print_result(int score) {
    if (is_passed(score)) {
```
