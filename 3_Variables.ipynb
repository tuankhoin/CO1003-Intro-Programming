### 2.1. Literal ki·ªÉu s·ªë nguy√™n

C h·ªó tr·ª£ nhi·ªÅu h·ªá c∆° s·ªë cho s·ªë nguy√™n:

| Chuy·ªÉn ƒë·ªïi            | Ph∆∞∆°ng ph√°p                                                   |

````python
%%bash
cat << 'EOF' > literals.c
#include <stdio.h>
int main() {
    printf("%d\n", 20);
    printf("%d\n", 020);
    printf("%d\n", 0x20);
    printf("======\n");
    printf("%d\n", 30);
    printf("10 x floor(30/8) + (30 mod 8) = %o\n", 30);
    printf("16 x floor(30/16) + (30 mod 16) = %X\n", 30);
    return 0;
}
EOF
gcc literals.c -o literals
./literals
````

## 11. Bi·ªÉu th·ª©c v√† to√°n t·ª≠

### 11.1. Bi·ªÉu th·ª©c

Bi·ªÉu th·ª©c g·ªìm:
- To√°n h·∫°ng (bi·∫øn, h·∫±ng, literal)
- To√°n t·ª≠ (`+`, `-`, `*`, `/`, ‚Ä¶)

### 11.2. C√°c lo·∫°i to√°n t·ª≠

- **To√°n t·ª≠ s·ªë h·ªçc**: `+ - * / %`
- **To√°n t·ª≠ g√°n**: `=`, `+=`, `-=`, ‚Ä¶
- **To√°n t·ª≠ so s√°nh**: `== != > < >= <=`
- **To√°n t·ª≠ lu·∫≠n l√Ω**: `&& || !`
- **To√°n t·ª≠ bitwise**: `& | ^ ~ << >>`
- **To√°n t·ª≠ ƒëi·ªÅu ki·ªán**: `condition ? value1 : value2`


````python
%%bash
cat << 'EOF' > bitwise_demo.c
#include <stdio.h>
int main() {
    char a = 5, b = 6;              // a=00000101, b=00000110
    unsigned char c;
    printf("%d\n", a & b);         // 4
    printf("%d\n", a | b);         // 7
    printf("%d\n", a ^ b);         // 3
    printf("%d\n", a << 2);        // 20
    printf("%d\n", b >> 2);        // 1
    printf("%d\n", ~a);            // -6 (signed char tr√™n nhi·ªÅu h·ªá)
    c = (unsigned char)(~a);
    printf("%u\n", c);             // 250
    return 0;
}
EOF
gcc bitwise_demo.c -o bitwise_demo
./bitwise_demo
````
