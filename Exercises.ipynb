# Đáp án

1. **B** — `x` bắt đầu từ 1, vào vòng lặp thì tăng trước rồi mới in nên ra `2 3 4`.
2. **C** — `switch` không có `break`, vào `case 2` rồi rơi tiếp xuống `case 3`.
3. **C** — `++a` làm `a = 5`, còn `b--` dùng giá trị 7 trước rồi mới giảm, nên `5 + 7 = 12`.
4. **B** — tổng `1 + 2 + 3 + 4 = 10`.
5. **D** — điều kiện `x > 5` đúng nên `y = 3 + 2 = 5`.
6. **B** — `*p = arr[0] = 5`, còn `*(p + 2) = arr[2] = 7`.
7. **B** — `a` truyền bằng giá trị nên không đổi, `b` truyền bằng địa chỉ nên tăng thành 22.
8. **C** — `f(3) = 3 + f(2) = 3 + 2 + f(1) = 3 + 2 + 1 = 6`.
9. **A** — truy cập đúng thành viên struct bằng dấu `.` nên in `101 25.5`.
10. **A** — `%06.2f` in số thực rộng 6 ký tự, có đệm `0`, nên `3.5` thành `003.50`.
11. **B** — `do...while` luôn chạy ít nhất 1 lần nên in `5`.
12. **B** — `continue` bỏ qua lần `i == 3`, nên in `1 2 4 5`.
13. **B** — vòng ngoài chạy 3 lần, vòng trong chạy 2 lần, tổng `3 × 2 = 6`.
14. **A** — mảng ký tự có `'\0'` ở cuối nên được in như chuỗi `"Hello"`.
15. **D** — `*(p + 1)` chính là `a[1]`, nên gán thành 10 và in ra 10.
16. **C** — hàm `square(x)` trả về `x * x`, nên `square(5) = 25`.
17. **A** — `3 > 2` đúng nhưng `5 < 4` sai, `true && false` cho kết quả `0`.
18. **B** — `9 % 2 = 1` nên không chẵn, in `"Odd"`.
19. **C** — `*p` là giá trị của `a`, tức 5, nên `*p + 3 = 8`.
20. **D** — `g(3) = 2 + g(2) = 2 + 2 + g(1) = 2 + 2 + 2 + g(0) = 6`.
